```python
import matplotlib.pyplot as plt


ages = [24, 29, 35, 36, 70, 18, 65, 45, 50]
salaries = [35, 42, 80, 72, 42, 25, 50, 110, 112]

# Plotting
plt.bar(ages, salaries)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Bar Graph of Age vs Salary')
plt.show()
```

```python

plt.plot(ages, salaries, marker='o')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Line Graph of Age vs Salary')
plt.grid(True)
plt.show()
```

```python

classes = ['A', 'A', 'B', 'A', 'C', 'C', 'A', 'B', 'B']
class_counts = {'A': 4, 'B': 3, 'C': 2}

plt.pie(class_counts.values(), labels=class_counts.keys(), autopct='%1.1f%%')
plt.title('Pie Chart of Class Distribution')
plt.show()
```

```python

plt.scatter(ages, salaries)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Age vs Salary')
plt.show()
```

```python

plt.hist(ages, bins=5, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Ages')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example data
ages = [24, 29, 35, 36, 70, 18, 65, 45, 50]
salaries = [35, 42, 80, 72, 42, 25, 50, 110, 112]

# Create a box plot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(data=[ages, salaries], orient='v')
plt.xticks([0, 1], ['Age', 'Salary'])
plt.title('Box Plot of Age and Salary')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
