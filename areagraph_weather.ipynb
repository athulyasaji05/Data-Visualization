```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('seattle-weather.csv')
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(14, 8))
```

```python
plt.fill_between(df['date'], df['precipitation'], alpha=0.5, label='Precipitation')
plt.fill_between(df['date'], df['temp_max'], alpha=0.5, label='Max Temperature')
plt.fill_between(df['date'], df['temp_min'], alpha=0.5, label='Min Temperature')
plt.fill_between(df['date'], df['wind'], alpha=0.5, label='Wind')
```
